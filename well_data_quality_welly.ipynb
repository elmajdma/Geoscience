# C-Create Keys with intersted curves

```python
all_curves_keys=['GR','RT','BS','NPHI','RACEHM','RACELM','RD','RHOB','ROP','RPCEHM','RPCELM',
             'ABDCQF01','ABDCQF02','ABDCQF03','ABDCQF04','CALI','DRHO','PEF','RM', 'DT','DTS',
             'NBGRCFM','CARB_FLAG','COAL_FLAG','ROP5_RM','ROP5_RM','RS','SAND_FLAG']
```

```python
selected_curves_keys=['GR','RT','BS','NPHI','RD','RHOB','CALI','DRHO','PEF', 'DT']
```

# D-Create curves QC Rules 

```python
qc_test = {'All':[wq.no_similarities],
           'Each':[wq.no_flat],  # checking if the curve contains constant values
                   #wq.no_gaps,   # checking for gaps
                   #wq.not_empty],# checking if the entire curve is not empty
           'GR': [
               wq.all_positive, # checking if all values are positive
               wq.all_between(0, 250), # checking if all values in some range
               wq.check_units(['API', 'GAPI']) # checking units
           ],
           'RT': [wq.all_positive, wq.all_between(0, 100000),wq.check_units(['API', 'GAPI'])],
           'BS':[wq.all_positive,wq.all_between(6,16),wq.check_units('inches')],
           'NPHI':[wq.all_positive,wq.all_between(45,-15)],
           'RHOB': [wq.all_positive,wq.all_between(1.5, 3),wq.check_units(['G/CC', 'g/cm3'])],
           'DT':[wq.all_positive,wq.all_between(40,140)],
           'ROP': [wq.all_positive,wq.all_between(0, 300),wq.check_units(['ft/hr', 'm/hr'])],
           'CALI': [wq.all_positive,wq.all_between(0, 300),wq.check_units(['ft/hr', 'm/hr'])],
           'PEF': [wq.all_positive],
           'DT': [wq.all_positive],
           }
```

```python
alias={
     "GR":["GR","NBGRCFM"],
     "RHOB":["RHOB","DEN"],
     "DT":["DT","DDT"],
     "RD":["RD","RDD"],
     "RT":["RT","RRT"],
     "NPHI":["NPH","NPHI"]
}
```

# E-Run Test

```python
HTML(volve_logs_qc.curve_table_html(keys=all_curves_keys, tests=qc_test))
```
